```python
# Ensure tpep_pickup_datetime is in datetime format
data['tpep_pickup_datetime'] = pd.to_datetime(data['tpep_pickup_datetime'])

# Round down to the nearest 30-minute interval
data['pickup_interval'] = data['tpep_pickup_datetime'].dt.floor('30T')

# Group by pickup_interval and calculate averages
interval_stats = data.groupby('pickup_interval').agg(
    average_passengers=('passenger_count', 'mean'),
    average_fare=('fare_amount', 'mean')
).reset_index()

# Display the results
print(interval_stats)

# Optionally, save the results to a CSV file
interval_stats.to_csv('interval_stats.csv', index=False)
```

```python
# Ensure tpep_pickup_datetime is in datetime format
data['tpep_pickup_datetime'] = pd.to_datetime(data['tpep_pickup_datetime'])

# Round down to the nearest 30-minute interval
data['pickup_interval'] = data['tpep_pickup_datetime'].dt.floor('30T')

# Group by pickup_interval and payment_type, and calculate the average fare amount
interval_payment_stats = data.groupby(['pickup_interval', 'payment_type']).agg(
    average_fare=('fare_amount', 'mean')
).reset_index()

# Find the interval with the maximum average fare for each payment type
max_fare_intervals = interval_payment_stats.loc[
    interval_payment_stats.groupby('payment_type')['average_fare'].idxmax()
].reset_index(drop=True)

# Display the results
print(max_fare_intervals)

# Optionally, save the results to a CSV file
max_fare_intervals.to_csv('max_fare_intervals.csv', index=False)
```

```python

# Ensure tpep_pickup_datetime is in datetime format
data['tpep_pickup_datetime'] = pd.to_datetime(data['tpep_pickup_datetime'])

# Round down to the nearest 30-minute interval
data['pickup_interval'] = data['tpep_pickup_datetime'].dt.floor('30T')

# Calculate the tip-to-fare ratio (avoid division by zero)
data['tip_to_fare_ratio'] = data['tip_amount'] / data['fare_amount']
data['tip_to_fare_ratio'].fillna(0, inplace=True)  # Handle NaN values
data['tip_to_fare_ratio'].replace([float('inf'), -float('inf')], 0, inplace=True)

# Group by pickup_interval and payment_type to calculate the average tip-to-fare ratio
interval_tip_stats = data.groupby(['pickup_interval', 'payment_type']).agg(
    average_tip_to_fare_ratio=('tip_to_fare_ratio', 'mean')
).reset_index()

# Find the interval with the maximum average tip-to-fare ratio for each payment type
max_tip_fare_intervals = interval_tip_stats.loc[
    interval_tip_stats.groupby('payment_type')['average_tip_to_fare_ratio'].idxmax()
].reset_index(drop=True)

# Display the results
print(max_tip_fare_intervals)

# Optionally, save the results to a CSV file
max_tip_fare_intervals.to_csv('max_tip_fare_intervals.csv', index=False)
```

```python
# Group by PULocationID to calculate the average fare amount
pickup_fare_stats = data.groupby('PULocationID').agg(
    average_fare=('fare_amount', 'mean')
).reset_index()

# Find the PULocationID with the highest average fare amount
highest_pickup_location = pickup_fare_stats.loc[
    pickup_fare_stats['average_fare'].idxmax()
]

# Group by DOLocationID to calculate the average fare amount
dropoff_fare_stats = data.groupby('DOLocationID').agg(
    average_fare=('fare_amount', 'mean')
).reset_index()

# Find the DOLocationID with the highest average fare amount
highest_dropoff_location = dropoff_fare_stats.loc[
    dropoff_fare_stats['average_fare'].idxmax()
]

# Display the results
print("Highest Pickup Location:")
print(highest_pickup_location)
print("\nHighest Dropoff Location:")
print(highest_dropoff_location)

# Optionally, save the results to a CSV file
pickup_fare_stats.to_csv('pickup_fare_stats.csv', index=False)
dropoff_fare_stats.to_csv('dropoff_fare_stats.csv', index=False)
```

```python
# Determine the 5 most common destinations for each pickup location
top_destinations = (
    data.groupby(['PULocationID', 'DOLocationID'])
    .size()
    .reset_index(name='trip_count')
    .sort_values(['PULocationID', 'trip_count'], ascending=[True, False])
    .groupby('PULocationID')
    .head(5)
)

# Filter the original data to keep only trips to the top 5 destinations for each pickup location
common = data.merge(
    top_destinations[['PULocationID', 'DOLocationID']],
    on=['PULocationID', 'DOLocationID']
)

# Display the resulting DataFrame
print(common)

# Optionally, save the resulting DataFrame to a CSV file
common.to_csv('common_trips.csv', index=False)
```

```python
# Ensure the tpep_pickup_datetime column is in datetime format
common['tpep_pickup_datetime'] = pd.to_datetime(common['tpep_pickup_datetime'])

# Round down to the nearest 30-minute interval
common['pickup_interval'] = common['tpep_pickup_datetime'].dt.floor('30T')

# Group by payment type and interval to calculate the average fare amount
average_fare_by_payment_and_interval = (
    common.groupby(['payment_type', 'pickup_interval'])
    .agg(average_fare=('fare_amount', 'mean'))
    .reset_index()
)

# Display the results
print(average_fare_by_payment_and_interval)

# Optionally, save the results to a CSV file
average_fare_by_payment_and_interval.to_csv('average_fare_by_payment_and_interval.csv', index=False)
```

```python
# Assuming 'average_fare_by_payment_and_interval' is the DataFrame for the 'common' DataFrame
# Assuming 'interval_payment_stats' is the DataFrame with the pre-computed stats

# Merge the two DataFrames on 'payment_type' and 'pickup_interval'
merged_stats = pd.merge(
    average_fare_by_payment_and_interval,
    interval_payment_stats,
    on=['payment_type', 'pickup_interval'],
    suffixes=('_common', '_overall')
)

# Compute the difference
merged_stats['fare_diff'] = (
    merged_stats['average_fare_common'] - merged_stats['average_fare_overall']
)

# Display the resulting DataFrame
print(merged_stats)

# Optionally, save the results to a CSV file
merged_stats.to_csv('fare_difference_stats.csv', index=False)
```

```python
# Ensure the previous merge is available as `merged_stats`

# Compute the ratio
merged_stats['fare_ratio'] = merged_stats['fare_diff'] / merged_stats['average_fare_overall']

# Display the resulting DataFrame
print(merged_stats)

# Optionally, save the results to a CSV file
merged_stats.to_csv('fare_ratio_stats.csv', index=False)
```

```python
# Convert datetime columns to datetime objects
data['tpep_pickup_datetime'] = pd.to_datetime(data['tpep_pickup_datetime'])
data['tpep_dropoff_datetime'] = pd.to_datetime(data['tpep_dropoff_datetime'])

# Sort the data by VendorID, tpep_dropoff_datetime, and PULocationID
data = data.sort_values(by=['VendorID', 'tpep_dropoff_datetime', 'PULocationID']).reset_index(drop=True)

# Initialize a 'chain' column with unique IDs
data['chain'] = range(len(data))

# Iterate through rows to find consecutive trips
for i in range(1, len(data)):
    current = data.iloc[i]
    previous = data.iloc[i - 1]

    if (
        current['VendorID'] == previous['VendorID'] and  # Same VendorID
        current['PULocationID'] == previous['DOLocationID'] and  # Matching locations
        current['tpep_pickup_datetime'] > previous['tpep_dropoff_datetime'] and  # Pickup after dropoff
        (current['tpep_pickup_datetime'] - previous['tpep_dropoff_datetime']).total_seconds() <= 120  # Within 2 minutes
    ):
        # Assign the same chain ID
        data.loc[i, 'chain'] = data.loc[i - 1, 'chain']

# Display the dataset with the 'chain' column
print(data)

# Save the result to a CSV file
data.to_csv('trip_chains.csv', index=False)
```
